## 进入 MMEditing 主目录

```python
import os
os.chdir('mmediting')
```

## 导入工具包

```python
import mmcv
from mmedit.edit import MMEdit

import matplotlib.pyplot as plt 
%matplotlib inline
```

## MMEdit实例

```python
editor = MMEdit(model_name='disco_diffusion')
```

## 输入Prompts文本

```python
text_prompts = {
    0: [
        'clouds surround the mountains and Chinese palaces,sunshine,lake,overlook,overlook,unreal engine,light effect,Dream，Greg Rutkowski,James Gurney,artstation'
    ]
}
```

## 指定输出保存路径

```python
out_path = 'outputs/text2img.jpg'
```

## 执行预测

```python
editor.infer(text=text_prompts, result_out_dir=out_path)
```

## 查看结果

```python
from PIL import Image
```
